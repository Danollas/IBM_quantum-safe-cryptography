```python
import cv2
import matplotlib.pyplot as plt
import cairosvg

# Convert SVG to PNG
svg_path = 'C:/Users/USER/Downloads/Windows_Logo_(1992-2001).svg'
png_path = 'temp_image.png'
cairosvg.svg2png(url=svg_path, write_to=png_path)

# Load the PNG image using OpenCV
image = cv2.imread(png_path, cv2.IMREAD_COLOR)

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Binarize the grayscale image
threshold = 127
_, binarized_image = cv2.threshold(gray_image, threshold, 255, cv2.THRESH_BINARY)

# Create a subplot with 3 columns for the original, grayscale, and binarized images
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Display the original image on the leftmost
axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[0].set_title('Original Image')
axes[0].axis('off')

# Display the grayscale image in the middle
axes[1].imshow(cv2.cvtColor(gray_image, cv2.COLOR_GRAY2RGB))
axes[1].set_title('Grayscale Image')
axes[1].axis('off')

# Display the binarized image on the rightmost
axes[2].imshow(cv2.cvtColor(binarized_image, cv2.COLOR_GRAY2RGB))
axes[2].set_title('Binarized Image')
axes[2].axis('off')

plt.tight_layout()
plt.show()
```
